1. Why modular code matters.

As your code gets longer, copy-pasting the same logic becomes difficult to maintain. Functions solve that problem.

````python
quiz_scores = [80, 90, 100]
````

````python
sum(quiz_scores), len(quiz_scores)
````

````python
quiz_average = sum(quiz_scores) / len(quiz_scores)

print(quiz_average)
````

````python
exam_scores = [70, 85, 95]
````

````python
exam_average = sum(exam_scores) / len(exam_scores)

print(exam_average)
````

Let's define a function so we can reuse the same code

`def` is the built-in Python keyword that tells Python we're making a function

````python
def calculate_mean(values):
    return sum(values) / len(values)
````

````python
# see our function in action with our quiz and exam scores
print(calculate_mean(quiz_scores))
print(calculate_mean(exam_scores))
````

````python
calculate_mean(quiz_scores) == quiz_average
````

Now the logic is written once and reused many times.

Why this matters in real work
Functions help you:

* reduce repetition
* debug more easily
* test one piece of logic at a time
* make code easier for teammates to read
* scale workflows across multiple datasets

### Key Terms
* Function definition: the code that creates the function
* Parameter: a variable listed in the function definition
* Argument: the actual value passed into the function
* Return value: the output the function sends back

````python
def multiply_numbers(a, b):
    result = a*b
    return result
````

````python
answer = multiply_numbers(4, 5)

answer
````

`a` and `b` are parameters

`4` and `5` are arguments

`20` is the return value

## 3. Important Function Concepts

---

### A. Functions Should Usually Do One Clear Thing

````python
def calculate_total(price, quantity):
    return price * quantity

calculate_total(100, 2)
````

### B. Default Parameters

If we don't specify an argument for a parameter, the default parameter is the argument

````python
def power_number(base, exponent=2):
    return base**exponent
````

````python
power_number(10)
````

````python
power_number(10, 3)
````

````python
power_number(exponent=2, base=10)
````

### C. Returning vs Printing

`return` sends a value back so it can be reused later.

````python
def add_numbers(a, b):
    return a+b

x = add_numbers(2, 2)
````

This prints the result, but does not return it for reuse.


````python
def add_numbers_print(a, b):
    print(a+b)

y = add_numbers_print(2, 2)
````

````python
x
````

````python
y
````

````python
print(calculate_mean(exam_scores))


calculate_mean(quiz_scores)
````

### D. Docstrings

Docstrings explain what a function does.

````python
def convert_to_upper(text):
    '''
    Look at this! I made a docstring!

    This function takes in some text and returns that text as uppercase
    
    '''
    return text.upper() # built-in python string method that upper cases strings
````

````python
convert_to_upper("hello")
````

````python
# python built-in help function provides the docstring of functions and other stuff
help(convert_to_upper)
````

````python
# help(print)
````

### E. Variable Scope

Variables created inside a function stay inside that function unless returned.

````python
def create_message(name):
    # message only exists as a variable INSIDE this function
    message = f"Welcome, {name}"
    return message

create_message("friend")
````

````python
# Now message has been defined OUTSIDE of our function and Python can see that variable
message = create_message("Paul Atreides")

message
````

## 6. Validation Inside Functions

Validation checks help catch problems early.

````python
def calculate_mean(values):
    return sum(values) / len(values)
````

### Why validation matters

Validation helps you:

- catch bad inputs early
- produce clearer error messages
- make functions more reliable
- prevent silent mistakes in analysis

````python
calculate_mean(scores)
````

````python
# Python throws an error, it doesn't know how to handle student as the function input!
calculate_mean(student)
````

````python
# re-write our calculate mean function with validation
def calculate_mean(values):
    if not isinstance(values, list):
        # we're checking to see that values is a list
        raise TypeError("values must be a list!")
        # we raise a TypeError, telling the user that the values input is not the right type
    if len(values) == 0:
        raise ValueError("values must not be an empty list!")
    return sum(values) / len(values)
````

## 7. Intro to Pandas for Function-Based Data Cleaning

Pandas is a Python library for working with tabular data.


````python
# !pip install pandas
````

````python
# import pandas, with the alias
import pandas as pd
````

````python
# define the dictionary
boring_dict = {
    "name": ["Ava", "Luis", "Mina", "Jordan"],
    "score": [95, None, 88, 91],
    "hours_studied": [10, 7, None, 8]
}
````

````python
# now convert it to a dataframe
df = pd.DataFrame(boring_dict)

print(df)
````

````python
# here are a few pandas methods to clean some data, then we'll put in a function
df.dropna()
````

````python
# .dropna() drops ANY missing values from a dataframe
# BE AWARE, YOU CAN LOSE A LOT OF DATA WITH THIS
# if we wanted to save the dataframe after dropping missing rows
# df = df.dropna()
````

````python
# we can fill missing values with .fillna()
# to access individual columns of a dataframe, df['name_of_col']
df['score'].fillna(df["score"].mean())

# took the score col, told pandas to fill in any missing data with the average score
````

````python
# Add a dervided column
# to create a new column, df['name_of_new_column'] = values_for_new_column
df["passed"] = df["score"] >= 70

df
````

````python
# Now we'll put them all into functions
def fill_missing_scores(df):
    df["score"] = df["score"].fillna(df["score"].mean())
    return df

fill_missing_scores(df)

df
````

## 8. Refactoring Repetitive Logic

One sign that you need a function is repeated code.

*Hint*: If you catch yourself writing the same logic 3+ times, that's a good indicator to refactor into a function!

````python
# take our scores from two classes

math_scores = [80, 90, 100]
science_scores = [75, 85, 95]
````

````python
# now manually calculate the average
math_average = sum(math_scores) / len(math_scores)
science_average = sum(science_scores) / len(science_scores)
````

````python
# Now refactor into a fucntion

def calculate_mean(values):
    return sum(values) / len(values)

math_average = calculate_mean(math_scores)
science_average = calculate_mean(science_scores)

math_average, science_average
````
